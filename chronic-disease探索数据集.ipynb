* Lets create a function to show all details and information about the dataset ËÆ©Êàë‰ª¨ÂàõÂª∫‰∏Ä‰∏™ÂáΩÊï∞Êù•ÊòæÁ§∫Êï∞ÊçÆÈõÜÁöÑÊâÄÊúâÁªÜËäÇÂíå‰ø°ÊÅØ

- [ ] featuresÔºöÂàóÂêç

- [ ] UniqueÔºöÂîØ‰∏ÄÂÄºÊï∞Èáè
- [ ] countÔºöÈùûÁ©∫ÂÄºÊï∞Èáè
- [ ] missedÔºöÁº∫Â§±ÂÄºÊï∞Èáè
- [ ] missed percentÔºöÁº∫Â§±ÂÄºÁôæÂàÜÊØî

minÔºöÊúÄÂ∞èÂÄº

```python
def df_details(dataset):
    column = []
    dtype = []
    unique = []
    count = []
    missing = []
    missing_perecent =[]
    mean_ = []
    std_ = []
    min_ = []
    max_ = []
    
    for col in df.columns:
        column.append(col)
        dtype.append(df[col].dtype)
        unique.append(len(df[col].unique()))
        count.append(len(df[col]))
        missing.append(df[col].isna().sum())
        missing_perecent.append((df[col].isna().sum()* 100)/len(df))
        if df[col].dtype == 'int64' or df[col].dtype == 'float64' :
            mean_.append(round(np.mean(df[col])))
            std_.append(round(np.std(df[col])))
            min_.append(round(np.min(df[col])))
            max_.append(round(np.max(df[col])))
        else:
            mean_.append('O')
            std_.append('O')
            min_.append('O')
            max_.append('O')
        
    datainfo = pd.DataFrame({'features' : column ,
                            'Dtype': dtype , 
                            'Unique' : unique,
                            'count': count,
                            'missed' : missing,
                            'missed percent' : missing_perecent ,                           
                            'mean':mean_,
                            'std':std_,
                            'max':max_,
                            'min':min_
                             })

    return datainfo 
```

```python
def show_year(col):
    sns.set_style('darkgrid')
    Year = df[col].value_counts().to_frame()
    Year['Year'] = Year.index
    Year.rename(columns={col:'count'},inplace=True)
    Year = Year.sort_values(by='Year' , ascending=True)
    plt.figure(figsize=(17 , 5) , dpi=300)
    plt.plot(Year.Year , Year['count'] , '--bo' ,label='line with marker' , c='k' ,linewidth='1.5')
    plt.title(col , c='b' , size =25)
    plt.xlabel('Year' , c='b' , size=15)
    plt.ylabel('Count' , c='b' , size=20)
    plt.xticks([2001,2007 , 2008 ,2009 , 2010 ,2011 , 2012 , 2013 , 2014 , 2015 , 2016 , 2017 , 2018 , 2019 , 2020, 2021]
               ,fontsize=15)
    plt.yticks(fontsize=20)

    plt.show()
```
