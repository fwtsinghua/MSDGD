# <p style="padding:10px;background-color:white;margin:0;color:black;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:900">Cleaning</p>

```python
full_missed_columns = ['Response' , 'StratificationCategory2','Stratification2' ,'StratificationCategory3',
                      'Stratification3' , 'ResponseID' , 'StratificationCategoryID2' , 'StratificationID2',
                      'StratificationCategoryID3','StratificationID3']
df.drop(columns=full_missed_columns , inplace=True)
```

<div style="font-family:Georgia;background-color:aliceblue; padding:30px; font-size:17px ; color:#034914">

**Description**<br>
* Now we have a dataset without full null columns  现在我们有一个没有完整空列的数据集
* Lets create a function to show all details and information about the dataset 让我们创建一个函数来显示数据集的所有细节和信息

- [ ] features：列名
- [ ] Dtype：数据类型
- [ ] Unique：唯一值数量
- [ ] count：非空值数量
- [ ] missed：缺失值数量
- [ ] missed percent：缺失值百分比
- [ ] mean：平均值
- [ ] std：标准差
- [ ] max：最大值
- [ ] min：最小值

      
min：最小值
</div>



```python
def df_details(dataset):
    column = []
    dtype = []
    unique = []
    count = []
    missing = []
    missing_perecent =[]
    mean_ = []
    std_ = []
    min_ = []
    max_ = []
    
    for col in df.columns:
        column.append(col)
        dtype.append(df[col].dtype)
        unique.append(len(df[col].unique()))
        count.append(len(df[col]))
        missing.append(df[col].isna().sum())
        missing_perecent.append((df[col].isna().sum()* 100)/len(df))
        if df[col].dtype == 'int64' or df[col].dtype == 'float64' :
            mean_.append(round(np.mean(df[col])))
            std_.append(round(np.std(df[col])))
            min_.append(round(np.min(df[col])))
            max_.append(round(np.max(df[col])))
        else:
            mean_.append('O')
            std_.append('O')
            min_.append('O')
            max_.append('O')
        
    datainfo = pd.DataFrame({'features' : column ,
                            'Dtype': dtype , 
                            'Unique' : unique,
                            'count': count,
                            'missed' : missing,
                            'missed percent' : missing_perecent ,                           
                            'mean':mean_,
                            'std':std_,
                            'max':max_,
                            'min':min_
                             })

    return datainfo 
```

```python
def show_year(col):
    sns.set_style('darkgrid')
    Year = df[col].value_counts().to_frame()
    Year['Year'] = Year.index
    Year.rename(columns={col:'count'},inplace=True)
    Year = Year.sort_values(by='Year' , ascending=True)
    plt.figure(figsize=(17 , 5) , dpi=300)
    plt.plot(Year.Year , Year['count'] , '--bo' ,label='line with marker' , c='k' ,linewidth='1.5')
    plt.title(col , c='b' , size =25)
    plt.xlabel('Year' , c='b' , size=15)
    plt.ylabel('Count' , c='b' , size=20)
    plt.xticks([2001,2007 , 2008 ,2009 , 2010 ,2011 , 2012 , 2013 , 2014 , 2015 , 2016 , 2017 , 2018 , 2019 , 2020, 2021]
               ,fontsize=15)
    plt.yticks(fontsize=20)

    plt.show()
```

```python
show_year('YearStart')
# print('--'*58)
show_year('YearEnd')
```
